```python
def last_day_of_month_in_timestamp(month):
    next_month = datetime.date(2020, month, 1).replace(day=28) + datetime.timedelta(days=4)
    date = (next_month - datetime.timedelta(days=next_month.day))
    element = datetime.datetime.strptime(str(date),"%Y-%m-%d")
    return int(datetime.datetime.timestamp(element))

def first_day_of_the_month_in_timestamp(month):
    date = datetime.datetime(2020, month, 1)
    element = datetime.datetime.strptime(str(date),"%Y-%m-%d %H:%M:%S")
    return int(datetime.datetime.timestamp(element))

def stock_dataframes(stock):
    table_list = []
    for month in range(1, 9):
        start_date = first_day_of_the_month_in_timestamp(month)
        end_date = last_day_of_month_in_timestamp(month)
        stock_data = f'https://finance.yahoo.com/quote/{stock}/history?period1={start_date}&period2={end_date}&interval=1d&filter=history&frequency=1d'
        tables = pd.read_html(stock_data)
        table_list.append(tables)
        
    return table_list
```

```python
netflix = stock_dataframes("NFLX")

jan_netflix_df = netflix[0][0]
feb_netflix_df = netflix[1][0]
mar_netflix_df = netflix[2][0]
apr_netflix_df = netflix[3][0]
may_netflix_df = netflix[4][0]
jun_netflix_df = netflix[5][0]
jul_netflix_df = netflix[6][0]
aug_netflix_df = netflix[7][0]
```

```python
netflix_df = [aug_netflix_df, jul_netflix_df, jun_netflix_df, may_netflix_df, apr_netflix_df, mar_netflix_df, feb_netflix_df, jan_netflix_df]
netflix_stock_df = pd.concat(netflix_df)
netflix_stock_df
```

```python
netflix_stock_df['Date']=pd.to_datetime(netflix_stock_df['Date'], errors='coerce').copy()
```

```python

netflix_stock_df.reset_index(inplace=True)
netflix_stock_df.drop(columns='index', inplace=True)
netflix_stock_df=netflix_stock_df[netflix_stock_df.Open!='*Close price adjusted for splits.**Adjusted close price adjusted for both dividends and splits.']
```

```python
netflix_stock_df=netflix_stock_df[['Date','Open','Close*']].copy()
netflix_stock_df.columns=['Date','netflix_open','netflix_close']
netflix_stock_df['netflix_open']=netflix_stock_df['netflix_open'].astype('str', copy=True).astype('float', copy=True)
netflix_stock_df['netflix_close']=netflix_stock_df['netflix_close'].astype('str', copy=True).astype('float', copy=True)
```

```python
netflix_stock_df['netflix_pert_change']=round(((netflix_stock_df['netflix_close']-netflix_stock_df['netflix_open'])/(netflix_stock_df['netflix_open']))*100,2)
netflix_stock_df.head()
```
